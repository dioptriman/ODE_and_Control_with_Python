```python
import control as ct
import numpy as np
import matplotlib.pyplot as plt
```

```python
k = 5
s = 6
G1 = ct.tf([1], [k,s])
```

```python
G1
```

```python
ct.dcgain(G1)
```

```python
G2 = ct.step_response(G1)
```

```python
t1, y1 = G2
plt.plot(t1, y1)
plt.show()
```

```python
ct.root_locus(G1)
```

```python
G3 = ct.tf([1],[1, 4, 5])
```

```python
resp=ct.root_locus_map(G3)
ct.root_locus_plot(resp)
```

```python
sys = ct.tf([1],[1, 2, 1], name='SISO transfer function')
response = ct.pole_zero_map(sys)
ct.pole_zero_plot(response)
```

```python
ct.bode_plot(sys)
```

```python
ct.nyquist_plot(sys)
```
