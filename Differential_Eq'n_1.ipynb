```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.arange(0., 100, 0.01)
print(t)
```

```python
model = 25 * np.exp(-0.056 * t) + 45
print(model)
```

```python
plt.plot(t, model)
plt.show()
```

```python
asymptote = 45 * np.ones(len(t))
print(asymptote)
```

```python
# Plot the model
plt.plot(t, model, 'b-', linewidth=1.5, label='Model')

# Plot the asymptote
plt.plot(t, asymptote, 'r--', linewidth=1.5, label='Asymptote')  # Dashed red line for the asymptote

# Set y-axis limits
plt.ylim(0, 80)  # Adjust limits to ensure the asymptote is visible

# Labels and title
plt.xlabel("Time")
plt.ylabel("Temperature")
plt.title("Temperature Model with Asymptote")
plt.legend(loc='best')
plt.grid(True)

# Show the plot
plt.show()
```
